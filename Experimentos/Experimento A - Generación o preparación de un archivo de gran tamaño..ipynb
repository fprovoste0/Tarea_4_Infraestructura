```python
import numpy as np
import pandas as pd

# Generador reproducible
rng = np.random.default_rng(42)

# Cantidad mínima solicitada
n_filas = 10_000_000

# Catálogos base
objetos = np.array([
    "Laptop", "Smartphone", "Tablet", "Audífonos", "Monitor",
    "Teclado", "Mouse", "Impresora", "Silla gamer", "Escritorio",
    "Camisa", "Pantalón", "Zapatos", "Reloj", "Mochila",
    "Cámara", "Bocina", "TV", "Consola", "Libro"
], dtype=object)

vendedores = np.array([
    "Ana", "Luis", "María", "Carlos", "Sofía",
    "Pedro", "Laura", "Jorge", "Elena", "Miguel"
], dtype=object)

# Columnas del dataset
id_usuario = rng.integers(100000, 999999, size=n_filas, dtype=np.int32)
objeto_idx = rng.integers(0, len(objetos), size=n_filas)
objeto = objetos[objeto_idx]

# Base de precios distinta por producto para que los valores varíen más entre categorías
precios_base_por_producto = np.array([
    900000, 1200000, 700000, 250000, 1800000,
    180000, 90000, 650000, 2200000, 1400000,
    45000, 55000, 80000, 300000, 75000,
    500000, 110000, 700000, 1500000, 60000
], dtype=np.int32)

# Ruido multiplicativo para mantener variación dentro de cada producto
ruido_precio = rng.lognormal(mean=0.0, sigma=0.35, size=n_filas)
precio_venta = np.clip(precios_base_por_producto[objeto_idx] * ruido_precio, 1000, None).round().astype(np.int32)

# Cantidad vendida por producto: artículos baratos tienden a venderse en más unidades
cantidad_promedio_por_producto = np.array([
    1, 1, 1, 2, 1,
    4, 6, 2, 1, 1,
    12, 10, 8, 3, 14,
    2, 5, 1, 1, 18
], dtype=np.int32)
cantidad_venta = rng.poisson(lam=cantidad_promedio_por_producto[objeto_idx]).astype(np.int32) + 1
vendedor = rng.choice(vendedores, size=n_filas)
canal_venta = rng.choice(["online", "presencial"], size=n_filas, p=[0.62, 0.38])

# Fechas aleatorias entre 2023 y 2026
inicio = np.datetime64("2023-01-01")
fin = np.datetime64("2026-12-31")
fecha_venta = inicio + rng.integers(0, (fin - inicio).astype(int) + 1, size=n_filas).astype("timedelta64[D]")

# Dataset final
dataset = pd.DataFrame({
    "id_usuario": id_usuario,
    "objeto": objeto,
    "precio_venta": precio_venta,
    "cantidad_venta": cantidad_venta,
    "vendedor": vendedor,
    "fecha_venta": fecha_venta,
    "canal_venta": canal_venta
})

# Vista rápida de verificación
print(dataset.shape)
display(dataset)
print(dataset.columns)

# Guardar el archivo en la carpeta dataset
dataset.to_csv("../dataset/dataset_tienda.csv", index=False)
print("Archivo guardado en ../dataset/dataset_tienda.csv")
```
